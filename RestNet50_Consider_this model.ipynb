```python

# Loading the ResNet50 model for feature extraction
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
query_image_folder = "/content/drive/MyDrive/competition_data/query"
query_image_paths = glob.glob(query_image_folder + "/*.jpg")
```

```python

gallery_image_folder = "/content/drive/MyDrive/competition_data/gallery"
gallery_image_paths = glob.glob(gallery_image_folder + "/*.jpg")
```

```python

query_features = []

for query_image_path in query_image_paths:
    input_image = Image.open(query_image_path).convert("RGB")  # Convert to RGB format#In order not to get error
    input_image = input_image.resize((224, 224))  # Resize the image
    input_array = np.array(input_image)
    input_array = np.expand_dims(input_array, axis=0)
    preprocessed_image = preprocess_input(input_array)

    features = base_model.predict(preprocessed_image)
    query_features.append(features)
```

```python
gallery_features = []

for gallery_image_path in gallery_image_paths:
    input_image = Image.open(gallery_image_path).convert("RGB")
    input_image = input_image.resize((224, 224))
    input_array = np.array(input_image)
    input_array = np.expand_dims(input_array, axis=0)
    preprocessed_image = preprocess_input(input_array)

    features = base_model.predict(preprocessed_image)
    gallery_features.append(features)
```

```python
query_features_array = np.array(query_features).reshape(len(query_features), -1)
gallery_features_array = np.array(gallery_features).reshape(len(gallery_features), -1)
```

```python
similarities = cosine_similarity(query_features_array, gallery_features_array)

for i, similarity_scores in enumerate(similarities):
    print(f"Similarities for query image {i+1}:")
    for j, score in enumerate(similarity_scores):
        print(f"Gallery image {j+1}: {score}")
    print()
```

```python

# Retrieve top N similar images for each query
N = 5  
top_similar_images = []
for similarity_scores in similarities:
    indices = np.argsort(similarity_scores)[::-1][:N]
    top_N_images = [gallery_image_paths[i] for i in indices]
    top_similar_images.append(top_N_images)
```

```python
# Print the top N similar images for each query
for i, query_image_path in enumerate(query_image_paths):
    print("Query Image:", query_image_path)
    print("Top Similar Images:")
    for j, similar_image_path in enumerate(top_similar_images[i]):
        print(f"{j+1}.", similar_image_path)
    print()
```

```python

# Calculate similarity scores between query and gallery images
similarity_scores = []

for query_index, query_feature in enumerate(query_features):
    query_image = query_image_paths[query_index]
    similarities = []

    for gallery_index, gallery_feature in enumerate(gallery_features):
        gallery_image = gallery_image_paths[gallery_index]
        similarity = cosine_similarity(
            np.reshape(query_feature, (1, -1)), np.reshape(gallery_feature, (1, -1))
        )[0][0]
        similarities.append((gallery_image, similarity))

    similarities.sort(key=lambda x: x[1], reverse=True)
    similarity_scores.append((query_image, similarities))
```

```python
# Print the top matching gallery images for each query image
for query_image, similarities in similarity_scores:
    print(f"Query Image: {query_image}")
    print("Top 1 Matches:")
    print(f"Gallery Image: {similarities[0][0]} - Similarity: {similarities[0][1]}")
    print("Top 5 Matches:")
    for gallery_image, similarity in similarities[:5]:
        print(f"Gallery Image: {gallery_image} - Similarity: {similarity}")
    print("Top 10 Matches:")
    for gallery_image, similarity in similarities[:10]:
        print(f"Gallery Image: {gallery_image} - Similarity: {similarity}")
    print()
```

```python

top_N = [1, 5, 10]  
res = dict()  

for i in range(len(query_image_paths)):
    similarities = cosine_similarity(query_features_array[i:i+1], gallery_features_array)
    indices = np.argsort(similarities)[0][::-1]
    top_N_indices = indices[:top_N[-1]]
    top_N_images = [os.path.basename(gallery_image_paths[j]) for j in top_N_indices]
    top_N_scores = similarities[0][top_N_indices]
    top_images = [(image, score) for image, score in zip(top_N_images, top_N_scores)]
    top_images_with_scores = [top_images[:N] for N in top_N]
    res[os.path.basename(query_image_paths[i])] = top_images_with_scores

# Print the results
for query_image, top_images_with_scores in res.items():
    print(f"Query Image: {query_image}")
    for i, N in enumerate(top_N):
        print(f"Top {N} Matches:")
        for j, (similar_image, similarity_score) in enumerate(top_images_with_scores[i]):
            print(f"{j+1}. {similar_image} - Similarity Score: {similarity_score}")
    print()
```
